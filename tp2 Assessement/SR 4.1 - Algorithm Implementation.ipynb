```python
# Import the pandas library
import pandas as pd

# Load dataset
data = pd.read_csv('C:\\Users\\Mr Maziya\\Documents\\Dataset1.csv')
```

```python
# Convert 'Date of Establishment' to datetime format, coercing errors to NaT
data['Date of Establishment'] = pd.to_datetime(data['Date of Establishment'], errors='coerce')

# Function to perform Selection Sort
def selection_sort(dataset, column, ascending=True):
    n = len(dataset)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if (dataset[column].iloc[j] < dataset[column].iloc[min_index]) if ascending else (dataset[column].iloc[j] > dataset[column].iloc[min_index]):
                min_index = j
        dataset.iloc[i], dataset.iloc[min_index] = dataset.iloc[min_index].copy(), dataset.iloc[i].copy()

# Choose the column to sort
column_to_sort = 'Date of Establishment'

# Sort the dataset in ascending order based on the chosen column
selection_sort(data, column_to_sort, ascending=True)

# Display the sorted dataset
print(data)
```
