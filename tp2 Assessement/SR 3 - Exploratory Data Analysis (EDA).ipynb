```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load dataset
data = pd.read_csv('C:\\Users\\Mr Maziya\\Documents\\Dataset1.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Province', data=data)
plt.title('Distribution of Settlements by Province')
plt.xlabel('Province')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
sanitation_counts = data['Sanitation'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(sanitation_counts, labels=sanitation_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Settlements with Sanitation Issues')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Longitude', y='Latitude', data=data, hue='Province', palette='Set2')
plt.title('Geographical Distribution of Settlements')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Province', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Province', y='Current Population', data=data)
plt.title('Distribution of Current Population Across Provinces')
plt.xlabel('Province')
plt.ylabel('Current Population')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=data, x='Longitude', y='Latitude', cmap='coolwarm', fill=True, thresh=0.05)
plt.title('KDE Plot of Latitude and Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
